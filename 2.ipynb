```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
import plotly
import seaborn as sns
```

```python
df = pd.read_csv('./datasets_top_4.csv')
df
```

```python
fig, ax = plt.subplots(figsize=(8, 10))

colores = ['#084594', '#2171b5', '#4292c6', '#CCCCCC']

def barrasapiladas(ax, datasets, categorias, colores):
    categorias.plot(kind='barh', stacked=True, ax=ax, color=colores, edgecolor='black')
    ax.set_yticklabels(datasets)
    ax.set_xlabel('Porcentaje')
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=4)

barrasapiladas(ax, df['dataset'], df[['cat1', 'cat2', 'cat3', 'otros']], colores)

fig.suptitle('Cómo se distribuyen las categorías dentro de los datasets?', fontsize=16)
ax.set_ylabel('Dataset')

plt.show()
```
