```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
import plotly
import seaborn as sns
```

```python
df = pd.read_csv('./viajes_surfistas.csv')
df
```

```python
country_counts = df['homecountry'].value_counts()

top_countries = ['United States of America', 'Japan', 'Australia', 'Spain']
df['homecountry'] = df['homecountry'].apply(lambda x: x if x in top_countries else 'Otro')
country_counts = df['homecountry'].value_counts()

fig, ax = plt.subplots(figsize=(8, 8))

ax.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=290)

ax.set_title('Porcentaje de surfistas por paÃ­s de procedencia')

plt.legend(title='Pais de Surfistas',labels=country_counts.index, loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```
